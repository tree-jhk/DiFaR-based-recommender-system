```python
import pandas as pd
```

```python
with open('./data/yahoo_movie_dataset/ydata-ymovies-mapping-to-movielens-v1_0.txt', 'r', encoding='ISO-8859-1') as file:
    ym_m = file.read()
ym_m = ym_m.split("\n")
```

```python
with open('./data/yahoo_movie_dataset/ydata-ymovies-user-demographics-v1_0.txt', 'r', encoding='ISO-8859-1') as file:
    ym_u = file.read()
ym_u = ym_u.split("\n")
```

```python
with open('./data/yahoo_movie_dataset/ydata-ymovies-user-movie-ratings-test-v1_0.txt', 'r', encoding='ISO-8859-1') as file:
    ym_test = file.read()
ym_test = ym_test.split("\n")
```

```python
with open('./data/yahoo_movie_dataset/ydata-ymovies-user-movie-ratings-train-v1_0.txt', 'r', encoding='ISO-8859-1') as file:
    ym_train = file.read()
ym_train = ym_train.split("\n")
```

```python
ym_r = ym_test + ym_train
```

```python
ym_m_dict = {"yahoo_movieId":list(), "title":list(), "movielens_movieId":list()}
ym_r_dict = {"userId":list(), "yahoo_movieId":list(), "yahoo_style_rating":list(), "rating":list()}
ym_u_dict = {"userId":list(), "birth_year":list(), "gender":list()}
ym_meta_dict = {"yahoo_movieId":list(), "title":list(), "synopsis":list(), "running_time":list(), "MPAA_rating":list(), "reason_MPAA_rating":list(), "release_date":list(), "release_date1":list(), "distributor":list(), "poster_url":list(), "genres":list(), "directors":list(), "director_ids":list(), "crew":list(), "crew_ids":list(), "crew_types":list(), "actors":list(), "actor_ids":list()}
```

```python
ym_m[0].split("\t")
```

```python
for val in ym_m:
    try:
        ymid, t, mid = val.split("\t")
        ym_m_dict["yahoo_movieId"].append(ymid)
        ym_m_dict["title"].append(t)
        ym_m_dict["movielens_movieId"].append(mid)
    except:
        print(val)
```

```python
ym_r[0].split("\t")
```

```python
for val in ym_r:
    try:
        uid, ymid, yr, r = val.split("\t")
        ym_r_dict["userId"].append(uid)
        ym_r_dict["yahoo_movieId"].append(ymid)
        ym_r_dict["yahoo_style_rating"].append(yr)
        ym_r_dict["rating"].append(r)
    except:
        print(val)
```

```python
for val in ym_u:
    try:
        uid, by, gen = val.split("\t")
        ym_u_dict["userId"].append(uid)
        ym_u_dict["birth_year"].append(by)
        ym_u_dict["gender"].append(gen)
    except:
        print(val)
```

```python
for val in ym_meta:
    try:
        
        ymid, t, s, rt, Mr, rMr, rd, rd1, dis, poster, g, d, did, c, cid, ct, a, aid = val.split("\t")
        ym_meta_dict["yahoo_movieId"].append(ymid)
        ym_meta_dict["title"].append(t)
        ym_meta_dict["synopsis"].append(s)
        ym_meta_dict["running_time"].append(rt)
        ym_meta_dict["MPAA_rating"].append(Mr)
        ym_meta_dict["reason_MPAA_rating"].append(rMr)
        ym_meta_dict["release_date"].append(rd)
        ym_meta_dict["release_date1"].append(rd1)
        ym_meta_dict["distributor"].append(dis)
        ym_meta_dict["poster_url"].append(poster)
        ym_meta_dict["genres"].append(g)
        ym_meta_dict["directors"].append(d)
        ym_meta_dict["director_ids"].append(did)
        ym_meta_dict["crew"].append(c)
        ym_meta_dict["crew_ids"].append(cid)
        ym_meta_dict["crew_types"].append(ct)
        ym_meta_dict["actors"].append(a)
        ym_meta_dict["actor_ids"].append(aid)
    except:
        print(val)
```

```python
ym_meta_df = pd.DataFrame(ym_meta_dict)
ym_u_df = pd.DataFrame(ym_u_dict)
ym_r_df = pd.DataFrame(ym_r_dict)
ym_m_df = pd.DataFrame(ym_m_dict)
```

```python
def f1(x:str):
    if "/I" in x:
        return x.replace("/I", "")
    else:
        return x

def f2(x:str):
    try:
        return str(int(f1(x).split()[-1].strip("\(").strip("\)")) // 10 * 10) + "'s"
    except:
        return ""
```

```python
df["release_year"] = df["title"].apply(lambda x:f2(f1(x)))
```

```python
movielens_tmdb_df = pd.read_csv("./data/movielens_tmdb.csv")
```

```python
movielens_tmdb_df
```

```python
cast_list = movielens_tmdb_df.cast.to_list()
director_list = movielens_tmdb_df.Director.to_list()
writer_list = movielens_tmdb_df.Writer.to_list()
movieId_list = movielens_tmdb_df.movieId.to_list()
```

```python
from collections import defaultdict
mw, md, mc = defaultdict(list), defaultdict(list), defaultdict(list)
```

```python
for m, w, d, c in zip(movieId_list, writer_list, director_list, cast_list):
    mw[m].append(w)
    md[m].append(d)
    mc[m].append(c)
```

```python
import numpy as np
np.NaN
```

```python
def f3(x, dic):
    a = dic[x]
    if np.NaN in a:
        return ""
    else:
        return a
```

```python
movielens_tmdb_df.sample(1)
```

```python
movielens_tmdb_df["writer"] = movielens_tmdb_df["movieId"].apply(lambda x:f3(x, mw))
movielens_tmdb_df.drop(labels=["Writer"], inplace=True, axis=1)
```

```python
movielens_tmdb_df["director"] = movielens_tmdb_df["movieId"].apply(lambda x:f3(x, md))
movielens_tmdb_df.drop(labels=["Director"], inplace=True, axis=1)
```

```python
movielens_tmdb_df["cast_"] = movielens_tmdb_df["movieId"].apply(lambda x:f3(x, mc))
movielens_tmdb_df.drop(labels=["cast"], inplace=True, axis=1)
```

```python
movielens_tmdb_df.drop(labels=["Original Story", "Original Film Writer"], inplace=True, axis=1)
```

```python
movielens_tmdb_df.rename(columns={"cast_":"cast"}, inplace=True)
```

```python
movielens_tmdb_df
```

```python
def f4(x):
    if isinstance(x, str):
        return ""
    else:
        x = list(set(x))
        return x[0]

movielens_tmdb_df["writer"] = movielens_tmdb_df["writer"].apply(lambda x:f4(x))
movielens_tmdb_df["director"] = movielens_tmdb_df["director"].apply(lambda x:f4(x))
movielens_tmdb_df["cast"] = movielens_tmdb_df["cast"].apply(lambda x:f4(x))
```

```python
movielens_tmdb_df_ = movielens_tmdb_df.drop_duplicates()
```
